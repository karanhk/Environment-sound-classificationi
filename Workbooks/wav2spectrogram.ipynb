# This workbook is to create a spectrogram from wav file and store it in form of png.

```python
import numpy as np
from scipy.io import wavfile
from scipy.signal import spectrogram
import matplotlib.pyplot as plt
import pandas as pd
```

## Path to wav and png

```python
path_wav = "/Users/karanhadiyal/Desktop/Environmental-sound-recognition/Highlights/sample.wav"
path_png = "/Users/karanhadiyal/Desktop/Environmental-sound-recognition/Highlights/"
```

## Read the wav

```python
sample_rate, audio_data = wavfile.read(path_wav)
f, t, Sxx = spectrogram(audio_data, fs=sample_rate)
```

## Plot the waveform and save png

```python
plt.figure(figsize=(10, 4))  # Adjust the figure size as needed
plt.plot(np.arange(len(audio_data)) / sample_rate, audio_data)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform of WAV File")
plt.savefig(path_png + "waveform.png")
```

## Plot the spectrogram and save png

```python
plt.pcolormesh(t, f, 10 * np.log10(Sxx))
plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")
plt.title("Spectrogram of WAV File")
plt.colorbar(label="Intensity (dB)")
plt.savefig(path_png + "spectrogram.png")
```
